```python
import numpy as np
import pandas as pd
```

#### Gerando dados:

```python

np.random.seed(42)

id = list(range(0,1000))
idade = list(np.random.randint(18, 90, size=1000))
sexo = list(np.random.choice(['F', 'M'], size=1000))
estados_civil = list(np.random.choice(['solteiro', 'casado', 'divorciado', 'viuvo'], size=1000)) 
sinistro = list(np.random.randint(0, 2, size=1000))
premio = list(np.round(np.random.uniform(500, 2000, size=1000), 2))

dados = {"id": id, "idade": idade, "sexo": sexo, "estados_civil": estados_civil, "sinistro": sinistro, "premio": premio}
```

#### Transformando os dados em um data.frame

```python
dataset = pd.DataFrame(dados)
dataset.head()
```

#### Visualizando os clientes que tem mais de 30 anos:

```python
dataset.loc[dataset['idade'] > 30]
```

#### Filtrando clientes sem sinistro com o método query 

```python
dataset.query('sinistro == 0')
```

##### Selecionando somente essas colunas do DataFrame

#### Criando uma coluna 'Categoria' no DataFrame para classificar os clientes de acordo com a idade

```python
def Categorizar_idade(idade):
    if idade >= 18 & idade <= 30:
        categoria = "Jovem"
    if idade >= 31 & idade <= 45:
        categoria = "Adulto"
    if idade > 45:
        categoria = "Sênior"

    return categoria

dataset['categoria'] = dataset['idade'].apply(Categorizar_idade)
dataset
```

#### Incluindo uma nova coluna chamada Desconto no DataFrame que calcula os seguintes descontos

- 10% sobre o prêmio para clientes Adultos e Seniors;
- 5% para Jovens;
- Considera que o desconto só é aplicado para clientes com sinistros igual a zero.

```python

def incluir_desconto(categoria):
    descontos = {"Jovem": 0.05, "Adulto": 0.1, "Sênior": 0.1}
    return descontos[categoria]

def cliente_sinistrado(sinistro):
    if sinistro == 0:
        multiplicador = 1
    else:
        multiplicador = 0
    
    return multiplicador
    

dataset['desconto'] = (dataset['categoria'].apply(incluir_desconto)) * (dataset['sinistro'].apply(cliente_sinistrado)) * (dataset['premio'])

dataset
```
