```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Importando os dados "clientes.csv" e criando um DataFrame 

```python
dados = pd.read_csv("C:\\Users\\rpbarros\\Documents\\Curso Alura\\Projeto_Pandas\\Dados\\clientes.csv", sep=";")
dados
```

#### Exibindo a informação de quantas linhas temos nessa tabela

```python
f"Quantidade de linhas: {dados.shape[0]}" 
```

#### Preenchendo os valores faltantes (NA) com zero na coluna quantidade de sinistros

```python
dados['Sinistros'].fillna(0, inplace=True)
dados
```

#### Ajustando as colunas "Sinistros", "Prêmio" e "Desconto" que possuem casas decimais desnecessárias:

- Na coluna de sinistros: ajustando para inteiro;
- Nas colunas "Prêmio" e "Desconto": ajustando para moeda, ou seja, duas casas decimais

```python
dados['Sinistros'] = dados['Sinistros'].astype(int)
dados['Prêmio'] = dados['Prêmio'].round(2)
dados['Desconto'] = dados['Desconto'].round(2)

dados
```

#### Removendo a coluna Desconto do DataFrame

```python
dados.pop('Desconto')

# ou podemos usar o metodo drop
# dados.drop(columns='Desconto', inplace= True)
dados
```

#### Ordenando os clientes pela idade

```python
dados.sort_values(by = 'Idade', inplace = True)
dados
```

#### Verificando o total de prêmio por estado civil

```python
grupo_estado_civil = dados.groupby('Estado Civil')

grupo_estado_civil[['Prêmio']].sum()
```

#### Contagem dos clientes por categoria (Jovem, Adulto, Sênior) com visualização do resultado em ordem crescente

```python
dados.Categoria.value_counts().sort_values()
```

#### Gráfico de barras para visualizar a contagem de clientes por estado civil

```python
qtd_estado_civil = dados['Estado Civil'].value_counts()

fig = qtd_estado_civil.plot.bar(color = '#BDD7EE')
fig.set_ylabel('Quantidade')
fig.set_title('Contagem de clientes por estado civil', {'fontsize': 22})
```

#### Gráfico de dispersão (scatter) para mostrar a relação entre a idade dos clientes e o valor do prêmio pago

```python
plt.scatter(dados.Idade, dados.Prêmio, color = '#9BC2E5')
plt.title('Idade X Prêmio')
plt.xlabel('Idade')
plt.ylabel('Prêmio')
```

#### Gráfico de pizza para visualizar a distribuição dos sinistros entre os diferentes sexos

```python
qtd_sexo = dados.Sexo.value_counts()

plt.pie(qtd_sexo, labels=qtd_sexo.index, colors = ['#FF97CB', '#9BC2E5'], autopct='%1.1f%%')
plt.title('Distribuição Sexos')

```

#### Histograma da quantidade de sinistros

```python
plt.hist(dados.Sinistros, color = ['#9BC2E5'])
plt.title('Distribuição quantidade de sinistros')

```
